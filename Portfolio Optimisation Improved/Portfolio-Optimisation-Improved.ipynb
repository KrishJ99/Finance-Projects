```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

```python
stocks = ['AAPL', 'MSFT', 'AMZN', 'NVDA', 'GOOGL', 'TSLA', 'BRK-B', 'META', 'V', 'JPM', 'COST', 'UNH', 'PEP', 'PFE', 'DAL', 'CL', 'MA', 'HD', 'NFLX', 'KO', 'PSA'] 

#Choosing stocks from a wide range of sectors;
# Technology, financial, healthcare, consumer goods, airlines, real estate ect.

start_date = '2015-05-01'
end_date = '2024-05-01'

stock_data = yf.download(stocks, start = start_date, end = end_date)

stock_data.head()
```

```python
adj_close = stock_data['Adj Close']

adj_close.head()
```

```python
# Save most recent year data such that we can compare

test_data = adj_close.iloc[adj_close.shape[0] - 252:,:] # Data set

train_data = adj_close.iloc[:adj_close.shape[0] - 252,:] # Training data for simulations

print("Testing data", test_data.shape)
print("Training data", train_data.shape)
```

### Returns


```python

expected_returns = test_data.pct_change() 
expected_returns.columns = stocks

expected_returns_a = pd.DataFrame(expected_returns.mean() * 250)  # Annualise average rate of return
expected_returns_a = expected_returns_a.T

DAR = test_data.pct_change().iloc[1:,:] + 1 # Daily average returns

GAR = pd.DataFrame(np.prod(DAR)**(1/float(6)) - 1)   # Geometric average rate of return = compound annual growth rate

annual_return = pd.concat([expected_returns_a.T, GAR], axis=1)
annual_return.columns = ["Average Arithmetic Returns", "Average Geometric Returns"]

print("Expected Annual Returns")
print(expected_returns_a)
print("\nDaily Average Returns")
print(DAR)
print("\nAnnual Returns")
print(annual_return)

```

- Sharpe Ratio < 1 , sub optimal

- Sharpe ratio > 1  = Acceptable

- Sharpe ratio > 2  = Very good

- Sharpe ratio > 3  = Excellent

```python
portfolio_returns = []
portfolio_volatility = []
sharpe_ratio = []

stock_weights = []

num_assets = len(stocks)
num_portfolios = 1000000   # Simulate 1 Million porfolios

daily_returns = test_data.pct_change()
annual_returns = (daily_returns.mean() * 250) + 1  # Annualised returns

cov_daily = daily_returns.cov()
cov_annual = cov_daily*250  # Annualised Covariance

print(annual_returns)
```

These are expected annual returns from last 8 years

```python
np.random.seed(3)

for i in range(num_portfolios):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)  # standardize the weights
    returns = (np.dot(weights, (annual_returns)))
    volatility = np.sqrt(np.dot(weights.T, np.dot(cov_annual, weights)))

    sharpe = ((returns - 1) / volatility)
    sharpe_ratio.append(sharpe)
    portfolio_returns.append(returns - 1)
    portfolio_volatility.append(volatility)
    stock_weights.append(weights)

'''
Sharpe ratio = calculates the risk adjusted return
- Suggests that adding assets to a portfolio that have low correlation can decrease porfolio risk 
without sacrificing returns
'''
```

```python
portfolio = {'Returns' : portfolio_returns,
             'volatility' : portfolio_volatility,
             'Sharpe_ratio' : sharpe_ratio}

for counter, symbol in enumerate(stocks):
    portfolio[symbol+'Weight'] = [Weight[counter] for Weight in stock_weights]

df = pd.DataFrame(portfolio)
df
```

#### Efficient frontier

```python

df.plot.scatter(x = 'volatility', y = 'Returns', c = 'Sharpe_ratio' ,
                cmap = 'plasma', grid = True, figsize = (10,8))

plt.title('Efficient Frontier')
plt.xlabel('Volatility')
plt.ylabel('Expected Returns')
plt.show()
```

This efficient frontier does not show the smooth, continuous curve typically associated with theoretical efficient frontiers. Instead, it shows a more realistic, granular representation based on  diverse stock portfolio of 21 stocks from many sectors and industries.
- As well it is simulated for 1,000,000 porfolios, so the plot is dense

#### Optimal porfolio

```python
min_volatility = df['volatility'].min()
max_sharpe = df['Sharpe_ratio'].max()

max_sharpe_port = df.loc[df['Sharpe_ratio'] == max_sharpe]
min_volatility_port = df.loc[df['volatility'] == min_volatility]

df.plot.scatter(x = 'volatility', y = 'Returns', c = 'Sharpe_ratio',
                cmap = 'plasma', grid = True, figsize = (10,8))

plt.scatter(x = max_sharpe_port['volatility'], y = max_sharpe_port['Returns'], c = 'red', marker = '*', s = 500, label = 'Max Sharpe')
plt.scatter(x = min_volatility_port['volatility'], y = min_volatility_port['Returns'], c = 'blue', marker = '*', s = 500, label = 'Min Volatility')
plt.title('Optimal Portfolio frontier')
plt.xlabel('Volatility')
plt.ylabel('Expected Returns')

plt.legend(loc = 'upper left')
plt.show()

```

```python
r_ef = pd.concat([min_volatility_port.T, max_sharpe_port.T], axis = 1)
r_ef.columns = ["Minimum Risk Portfolio (Adjusted)", "Max Sharpe Ratio Portfolio (Adjusted)"]
r_ef
```
